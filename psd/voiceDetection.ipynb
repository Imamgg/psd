# Voice detection multi label

```python
import os
import pandas as pd
import tsfel
import librosa
```

```python
base_path = "./dataset/voice"

cfg = tsfel.get_features_by_domain()

all_features = []

for person in os.listdir(base_path):
    person_path = os.path.join(base_path, person)
    if not os.path.isdir(person_path):
        continue

    for action in os.listdir(person_path):
        action_path = os.path.join(person_path, action)
        if not os.path.isdir(action_path):
            continue

        for filename in os.listdir(action_path):
            if filename.endswith(".mp3") or filename.endswith(".wav"):
                file_path = os.path.join(action_path, filename)

                try:
                    signal, sr = librosa.load(file_path, sr=None)
                    features = tsfel.time_series_features_extractor(cfg, signal, verbose=0)

                    features['label_orang'] = person
                    features['label_aksi'] = action
                    features['file'] = filename

                    all_features.append(features)

                    print(f"✅ Selesai ekstraksi: {file_path}")
                except Exception as e:
                    print(f"❌ Gagal memproses {file_path}: {e}")

final_df = pd.concat(all_features, ignore_index=True)

os.makedirs("data", exist_ok=True)

final_df.to_csv("data/features_tsfel_all_with_person_label.csv", index=False)
print("✅ Semua fitur berhasil disimpan ke 'data/features_tsfel_all_with_person_label.csv'")
print(f"Total samples: {len(final_df)}")
```

## ⚠️ PENTING: Re-extract Features dengan Sample Rate Konsisten

Masalah yang ditemukan:
- Audio dari Imam menggunakan sample rate 44100 Hz
- Audio dari Mufid menggunakan sample rate 16000 Hz
- Perbedaan ini menyebabkan fitur TSFEL yang diekstrak berbeda
- Model menjadi bias terhadap sample rate, bukan terhadap suara

**Solusi**: Re-extract semua fitur dengan **sample rate yang sama (22050 Hz)**

```python
# Re-extract features dengan sample rate konsisten
base_path = "./dataset/voice"

cfg = tsfel.get_features_by_domain()

# PENTING: Gunakan sample rate yang TETAP untuk semua audio
FIXED_SAMPLE_RATE = 22050

all_features = []

for person in os.listdir(base_path):
    person_path = os.path.join(base_path, person)
    if not os.path.isdir(person_path):
        continue

    for action in os.listdir(person_path):
        action_path = os.path.join(person_path, action)
        if not os.path.isdir(action_path):
            continue

        for filename in os.listdir(action_path):
            if filename.endswith(".mp3") or filename.endswith(".wav"):
                file_path = os.path.join(action_path, filename)

                try:
                    # Load dengan sample rate yang FIXED (bukan sr=None)
                    signal, sr = librosa.load(file_path, sr=FIXED_SAMPLE_RATE)
                    
                    print(f"📊 {file_path}")
                    print(f"   Original SR: native → Resampled to: {sr} Hz")
                    
                    features = tsfel.time_series_features_extractor(cfg, signal, verbose=0)

                    features['label_orang'] = person
                    features['label_aksi'] = action
                    features['file'] = filename

                    all_features.append(features)

                    print(f"✅ Selesai ekstraksi: {file_path}")
                except Exception as e:
                    print(f"❌ Gagal memproses {file_path}: {e}")

final_df = pd.concat(all_features, ignore_index=True)

# Buat direktori jika belum ada
os.makedirs("data", exist_ok=True)

final_df.to_csv("data/features_tsfel_fixed_sr_22050.csv", index=False)
print(f"\n✅ Semua fitur berhasil disimpan ke 'data/features_tsfel_fixed_sr_22050.csv'")
print(f"📊 Total samples: {len(final_df)}")
print(f"🎯 Sample rate yang digunakan: {FIXED_SAMPLE_RATE} Hz (KONSISTEN untuk semua audio)")
```

```python
data = "data/features_tsfel_all_with_person_label.csv"

df = pd.read_csv(data)

print("✅ Data berhasil dibaca!")
print("Jumlah sample:", len(df))
print("Jumlah kolom :", int(len(df.columns)) - 3)
```

```python
df.head()
```

```python
# Cek missing values dan infinite values
missing = df.isnull().sum()
print("Missing values:")
print(missing[missing > 0])

# Cek infinite values
import numpy as np
inf_cols = []
for col in df.select_dtypes(include=[np.number]).columns:
    if np.isinf(df[col]).any():
        inf_cols.append(col)

if inf_cols:
    print(f"\n⚠️ Kolom dengan nilai infinite: {len(inf_cols)}")
    print("Contoh kolom:", inf_cols[:5])
else:
    print("\n✅ Tidak ada nilai infinite")
```

```python
X = df.drop(columns=["label_aksi", "label_orang", "file"])
y_action = df["label_aksi"]
y_person = df["label_orang"]

# Handle NaN dan Infinite values
print(f"Shape sebelum cleaning: {X.shape}")

# Replace infinite dengan NaN
X = X.replace([np.inf, -np.inf], np.nan)

# Drop kolom yang semua nilainya NaN
X = X.dropna(axis=1, how='all')

# Fill NaN dengan median per kolom
X = X.fillna(X.median())

print(f"Shape setelah cleaning: {X.shape}")
print(f"✅ Data berhasil dibersihkan dari NaN dan Inf")
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder_action = LabelEncoder()
label_encoder_person = LabelEncoder()

y_action_encoded = label_encoder_action.fit_transform(y_action)
y_person_encoded = label_encoder_person.fit_transform(y_person)

print(label_encoder_action.classes_)
print(label_encoder_person.classes_)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train_a, X_test_a, y_train_a, y_test_a = train_test_split(
    X_scaled, y_action_encoded, test_size=0.2, random_state=42, stratify=y_action_encoded
)

X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(
    X_scaled, y_person_encoded, test_size=0.2, stratify=y_person_encoded
)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

#### Menggunakan semua fitur

```python
model_action = RandomForestClassifier(n_estimators=200, random_state=42)
model_action.fit(X_train_a, y_train_a)
y_pred_action = model_action.predict(X_test_a)

print("\n===== Model: LABEL_AKSI =====")
print(f"Akurasi: {accuracy_score(y_test_a, y_pred_action):.4f}")
print(confusion_matrix(y_test_a, y_pred_action))
print(classification_report(y_test_a, y_pred_action, target_names=label_encoder_action.classes_))
```

```python
model_person = RandomForestClassifier(n_estimators=200, random_state=42)
model_person.fit(X_train_b, y_train_b)
y_pred_action = model_person.predict(X_test_b)

print("\n===== Model: LABEL_ORANG =====")
print(f"Akurasi: {accuracy_score(y_test_b, y_pred_action):.4f}")
print(confusion_matrix(y_test_b, y_pred_action))
print(classification_report(y_test_b, y_pred_action, target_names=label_encoder_person.classes_))
```

#### Feature selection

```python
threshold = 0.3

selected_features_action = mutual_information_df_action[mutual_information_df_action['MI_Score'] > threshold]['Feature'].tolist()
selected_features_person = mutual_information_df_person[mutual_information_df_person['MI_Score'] > threshold]['Feature'].tolist()

# Ambil minimal 10 fitur terbaik jika hasil threshold terlalu sedikit
if len(selected_features_action) < 10:
    selected_features_action = mutual_information_df_action.head(10)['Feature'].tolist()

if len(selected_features_person) < 10:
    selected_features_person = mutual_information_df_person.head(10)['Feature'].tolist()

X_selected_action = X[selected_features_action]
X_selected_person = X[selected_features_person]

# PENTING: Buat scaler baru untuk masing-masing
scaler_action = StandardScaler()
scaler_person = StandardScaler()

X_selected_action_scaled = scaler_action.fit_transform(X_selected_action)
X_selected_person_scaled = scaler_person.fit_transform(X_selected_person)

print(f"✅ Jumlah fitur untuk label aksi: {len(selected_features_action)}")
print(f"✅ Jumlah fitur untuk label orang: {len(selected_features_person)}")
print(f"\nTop 5 fitur untuk AKSI:")
for i, feat in enumerate(selected_features_action[:5], 1):
    print(f"  {i}. {feat}")

print(f"\nTop 5 fitur untuk ORANG:")
for i, feat in enumerate(selected_features_person[:5], 1):
    print(f"  {i}. {feat}")
```

```python
model_action = RandomForestClassifier(n_estimators=200, random_state=42)
model_action.fit(X_train_a, y_train_a)

y_pred_action = model_action.predict(X_test_a)

print("=== Evaluasi Model untuk LABEL_AKSI ===")
print(classification_report(y_test_a, y_pred_action))
print("Akurasi:", accuracy_score(y_test_a, y_pred_action))

plt.figure(figsize=(5,4))
sns.heatmap(confusion_matrix(y_test_a, y_pred_action), annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix - LABEL_AKSI")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

#### Test Prediksi dengan Audio Baru

```python
def predict_voice(audio_path):
    """
    Fungsi untuk memprediksi voice dari file audio baru
    
    Parameters:
    -----------
    audio_path : str
        Path ke file audio (.wav atau .mp3)
    
    Returns:
    --------
    dict : Hasil prediksi dengan informasi detail
    """
    # Load model dan preprocessing objects
    with open('model/model_action.pkl', 'rb') as f:
        model_action = pickle.load(f)
    
    with open('model/model_person.pkl', 'rb') as f:
        model_person = pickle.load(f)
    
    with open('model/scaler_action.pkl', 'rb') as f:
        scaler_action = pickle.load(f)
    
    with open('model/scaler_person.pkl', 'rb') as f:
        scaler_person = pickle.load(f)
    
    with open('model/selected_features_action.pkl', 'rb') as f:
        selected_features_action = pickle.load(f)
    
    with open('model/selected_features_person.pkl', 'rb') as f:
        selected_features_person = pickle.load(f)
    
    with open('model/tsfel_config.pkl', 'rb') as f:
        cfg = pickle.load(f)
    
    # Load dan ekstraksi fitur dari audio
    signal, sr = librosa.load(audio_path, sr=None)
    features = tsfel.time_series_features_extractor(cfg, signal, verbose=0)
    
    # Handle NaN dan Inf
    features = features.replace([np.inf, -np.inf], np.nan)
    features = features.fillna(features.median())
    
    # Prediksi AKSI (buka/tutup)
    X_action = features[selected_features_action]
    X_action_scaled = scaler_action.transform(X_action)
    pred_action = model_action.predict(X_action_scaled)[0]
    pred_action_proba = model_action.predict_proba(X_action_scaled)[0]
    
    # Prediksi ORANG (imam/mufid)
    X_person = features[selected_features_person]
    X_person_scaled = scaler_person.transform(X_person)
    pred_person = model_person.predict(X_person_scaled)[0]
    pred_person_proba = model_person.predict_proba(X_person_scaled)[0]
    
    return {
        'file': audio_path,
        'prediksi_aksi': pred_action,
        'confidence_aksi': max(pred_action_proba) * 100,
        'prediksi_orang': pred_person,
        'confidence_orang': max(pred_person_proba) * 100,
        'probabilitas_aksi': dict(zip(model_action.classes_, pred_action_proba)),
        'probabilitas_orang': dict(zip(model_person.classes_, pred_person_proba))
    }

# Test dengan beberapa file audio
print("🎯 Testing Model dengan Audio dari Dataset\n")

test_files = [
    "./dataset/voice/imam/buka/buka1.wav",
    "./dataset/voice/imam/tutup/tutup1.wav",
    "./dataset/voice/mufid/buka/Record (online-voice-recorder.com) (1).mp3",
    "./dataset/voice/mufid/tutup/Record (online-voice-recorder.com) (1).mp3"
]

for audio_file in test_files:
    if os.path.exists(audio_file):
        result = predict_voice(audio_file)
        print(f"📁 File: {os.path.basename(audio_file)}")
        print(f"   Prediksi Aksi: {result['prediksi_aksi']} (confidence: {result['confidence_aksi']:.2f}%)")
        print(f"   Prediksi Orang: {result['prediksi_orang']} (confidence: {result['confidence_orang']:.2f}%)")
        print(f"   Detail probabilitas aksi: {result['probabilitas_aksi']}")
        print(f"   Detail probabilitas orang: {result['probabilitas_orang']}")
        print()
```

#### Cross-Validation untuk Evaluasi Model yang Lebih Robust

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold

# Cross-validation untuk model AKSI
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_scores_action = cross_val_score(
    RandomForestClassifier(n_estimators=200, random_state=42),
    X_selected_action_scaled,
    y_action,
    cv=cv,
    scoring='accuracy'
)

print("=== Cross-Validation: Model AKSI ===")
print(f"Scores per fold: {cv_scores_action}")
print(f"Mean Accuracy: {cv_scores_action.mean():.4f} (+/- {cv_scores_action.std():.4f})")

# Cross-validation untuk model ORANG
cv_scores_person = cross_val_score(
    RandomForestClassifier(n_estimators=200, random_state=42),
    X_selected_person_scaled,
    y_person,
    cv=cv,
    scoring='accuracy'
)

print("\n=== Cross-Validation: Model ORANG ===")
print(f"Scores per fold: {cv_scores_person}")
print(f"Mean Accuracy: {cv_scores_person.mean():.4f} (+/- {cv_scores_person.std():.4f})")

# Visualisasi CV scores
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].bar(range(1, 6), cv_scores_action, color='skyblue', edgecolor='navy')
axes[0].axhline(y=cv_scores_action.mean(), color='red', linestyle='--', label=f'Mean: {cv_scores_action.mean():.4f}')
axes[0].set_xlabel('Fold')
axes[0].set_ylabel('Accuracy')
axes[0].set_title('Cross-Validation Scores - Model AKSI')
axes[0].set_ylim([0, 1])
axes[0].legend()
axes[0].grid(axis='y', alpha=0.3)

axes[1].bar(range(1, 6), cv_scores_person, color='lightcoral', edgecolor='darkred')
axes[1].axhline(y=cv_scores_person.mean(), color='blue', linestyle='--', label=f'Mean: {cv_scores_person.mean():.4f}')
axes[1].set_xlabel('Fold')
axes[1].set_ylabel('Accuracy')
axes[1].set_title('Cross-Validation Scores - Model ORANG')
axes[1].set_ylim([0, 1])
axes[1].legend()
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```
