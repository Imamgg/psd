# Data Preprocessing

## Definisi & Tujuan
* Data preprocessing adalah proses membersihkan dan mengubah data mentah ke dalam format yang dapat digunakan dan sesuai untuk analisis atau penambangan data (data mining).
* Tujuannya adalah meningkatkan kualitas data, sehingga hasil analisis lebih akurat dan dapat dipercaya.
* Dalam prosesnya, masalah seperti nilai hilang (missing values), duplikasi, dan inkonsistensi diatasi.

---

## Langkah-Langkah / Tahapan

Artikel menyebutkan beberapa tahapan kunci dalam data preprocessing sebagai berikut:

| Tahap                                       | Penjelasan Singkat                                                                                                                                  |
| ------------------------------------------- | --------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Data Cleaning (Pembersihan Data)**        | Mengidentifikasi dan memperbaiki kesalahan atau inkonsistensi: menangani nilai hilang, data noise, dan penghapusan duplikasi.   |
| **Data Integration (Integrasi Data)**       | Menggabungkan data dari sumber berbeda menjadi satu dataset yang konsisten.                                                     |
| **Data Transformation (Transformasi Data)** | Mengonversi data ke format yang lebih sesuai: normalisasi, standarisasi, diskretisasi, agregasi, dan hierarki konsep.           |
| **Data Reduction (Reduksi Data)**           | Memperkecil ukuran dataset sambil mempertahankan informasi penting: seleksi fitur, ekstraksi fitur, reduksi dimensi, kompresi.  |

---

## Kegunaan & Kekurangan

### Kegunaan (Keuntungan)

Proses preprocessing berguna dalam berbagai konteks, menurut artikel tersebut:

* Data Warehousing — menyatukan, membersihkan, dan menyusun data sebelum disimpan.
* Data Mining — membuat data siap untuk algoritma penambangan data.
* Machine Learning & Deep Learning — mempersiapkan data untuk pelatihan model (misalnya normalisasi, encoding).
* Web Mining, Business Intelligence, dan aplikasi lain juga membutuhkan preprocessing agar data lebih representatif dan bersih.

### Kekurangan / Tantangan

Beberapa kelemahan yang disebutkan:

* Memerlukan waktu dan usaha yang besar (time-consuming).
* Membutuhkan sumber daya (komputasi, keahlian).
* Ada risiko kehilangan data berharga jika preprocessing dilakukan secara keliru.
* Kompleksitas meningkat ketika dataset besar atau sangat heterogen.
