# Business Understanding

**Business Understanding** adalah fase fondasi dari setiap proyek, terutama dalam analisis data, di mana Anda berusaha memahami **konteks, tujuan, dan kebutuhan bisnis** di balik sebuah permintaan.
## Komponen Inti dari Business Understanding

Untuk membangun pemahaman bisnis yang solid, Anda perlu menguraikan beberapa elemen kunci. Ini sering kali menjadi bagian pertama dari metodologi standar seperti CRISP-DM (*Cross-Industry Standard Process for Data Mining*).

1.  **Menentukan Tujuan Bisnis (Determine Business Objectives)**
    * **Apa masalah utama atau peluang yang ingin dikejar oleh perusahaan?** Ini adalah gambaran besarnya. Tujuannya harus dijelaskan dari sudut pandang bisnis, bukan teknis.
    * *Contoh buruk:* "Membangun model machine learning untuk prediksi."
    * *Contoh baik:* "Meningkatkan profitabilitas perusahaan dengan mengurangi jumlah pelanggan yang berhenti berlangganan (churn) sebesar 15% dalam 6 bulan ke depan."

2.  **Menilai Situasi Saat Ini (Assess Situation)**
    * **Apa saja yang sudah diketahui tentang masalah ini?** Apa yang sudah pernah dicoba sebelumnya dan bagaimana hasilnya?
    * **Sumber daya apa yang kita miliki?** (Data, orang, tools, budget).
    * **Apa saja batasan (constraints) yang ada?** (Waktu, anggaran, regulasi privasi data, batasan teknis).
    * **Apa saja risiko yang mungkin terjadi?** (Data tidak memadai, hasil tidak signifikan, penolakan dari stakeholder).

3.  **Menentukan Tujuan Proyek/Analisis (Determine Project Goals)**
    * **Bagaimana proyek ini akan membantu mencapai tujuan bisnis di atas?** Ini menerjemahkan tujuan bisnis yang luas menjadi tujuan teknis yang lebih spesifik.
    * *Contoh:* "Mengidentifikasi faktor-faktor utama yang menyebabkan pelanggan churn dan membangun model prediksi untuk mengenali pelanggan yang berisiko tinggi akan churn bulan depan."

4.  **Menetapkan Kriteria Sukses (Define Success Criteria)**
    * **Bagaimana kita tahu bahwa proyek ini berhasil?** Kriteria ini harus spesifik dan terukur.
    * **Kriteria Sukses Bisnis:** "Tingkat churn menurun sebesar 15% setelah intervensi berdasarkan hasil model diterapkan selama satu kuartal."
    * **Kriteria Sukses Teknis:** "Model prediksi yang dibangun memiliki akurasi di atas 85% dan mampu mengidentifikasi setidaknya 70% dari pelanggan yang benar-benar akan churn (recall)."
