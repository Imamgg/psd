## Data Understanding

**Data Understanding** adalah fase di mana Anda melakukan investigasi awal terhadap data untuk membiasakan diri, mengidentifikasi masalah kualitas, menemukan wawasan pertama, dan/atau membentuk hipotesis. Fase ini adalah jembatan krusial antara tujuan bisnis yang abstrak dengan persiapan data teknis.

### Aktivitas Inti dalam Data Understanding

Fase ini terdiri dari empat kegiatan utama yang saling berhubungan:

1.  **Pengumpulan Data Awal (Initial Data Collection)**
    * **Tugas:** Mengumpulkan data yang relevan dari berbagai sumber yang telah diidentifikasi. Ini bisa berupa tabel dari database, file CSV, akses API, atau data tidak terstruktur seperti teks dan gambar.
    * **Pertanyaan Kunci:** Apakah kita bisa mengakses data yang dibutuhkan? Siapa penanggung jawab data ini? Bagaimana cara mendapatkannya?

2.  **Deskripsi Data (Describe Data)**
    * **Tugas:** Mendokumentasikan properti dasar dari data yang diperoleh. Ini seperti membuat "kartu identitas" untuk setiap set data.
    * **Apa yang diperiksa:**
        * **Jumlah data:** Berapa banyak baris (observasi) dan kolom (fitur/atribut)?
        * **Tipe data:** Apakah kolom ini angka (integer, float), teks (string), kategori, atau tanggal?
        * **Definisi kolom:** Apa arti dari setiap kolom? `trx_amt` itu jumlah transaksi atau nilai transaksi? Satuannya apa?
    * **Hasil:** Laporan deskripsi data yang memberikan gambaran umum tentang format dan skala data.

3.  **Eksplorasi Data (Explore Data / EDA - Exploratory Data Analysis)**
    * **Tugas:** Ini adalah bagian investigasi yang paling mendalam. Anda "menggali" data untuk menemukan pola, anomali, hubungan, dan wawasan awal. Di sinilah hipotesis mulai terbentuk.
    * **Teknik yang umum digunakan:**
        * **Statistik Deskriptif:** Menghitung nilai-nilai seperti rata-rata (mean), nilai tengah (median), modus, standar deviasi, nilai minimum & maksimum.
        * **Visualisasi:** Membuat grafik untuk memahami distribusi dan hubungan antar variabel. Contohnya:
            * **Histogram/Box Plot:** Untuk melihat sebaran nilai dari satu variabel numerik (misal: persebaran umur pelanggan).
            * **Bar Chart:** Untuk melihat frekuensi dari variabel kategori (misal: jumlah pelanggan per kota).
            * **Scatter Plot:** Untuk melihat hubungan antara dua variabel numerik (misal: apakah ada hubungan antara pengeluaran iklan dan penjualan?).
            * **Heatmap:** Untuk melihat korelasi antara banyak variabel sekaligus.

4.  **Verifikasi Kualitas Data (Verify Data Quality)**
    * **Tugas:** Menilai apakah data yang ada "cukup baik" untuk analisis. Data di dunia nyata jarang sekali sempurna.
    * **Masalah yang dicari:**
        * **Missing Values:** Apakah ada data yang hilang atau kosong? Seberapa banyak?
        * **Inkonsistensi:** Apakah ada data yang tidak konsisten? (Contoh: "SBY", "Surabaya", "Sby" di kolom kota).
        * **Outliers:** Apakah ada nilai yang ekstrem atau tidak masuk akal? (Contoh: umur pelanggan 200 tahun).
        * **Duplikasi:** Apakah ada baris data yang sama persis?
